```python
## import all libraries

import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
Z = linkage(x,'average')
```

```python
# calculate full dendogram

import matplotlib.pyplot as plt

plt.figure(figsize=(25,10))
plt.xlabel('Sample index')
plt.ylabel('distance')
dendrogram(Z,leaf_rotation=90,leaf_font_size=8)
plt.show()

## Here if create any horizontal line then maximum 2 cluster will made i.e. of the longest distance horizontal line
```

```python
## Another case study using AgglomerativeClustering

df = pd.read_csv('shopping_data.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# if selecting like [:,3:5] # it will return column anullan income and spending score

data = df.iloc[:,3:5] # if we selected only [3:5] then output is 4 and 5th row only
data
```

```python
data.values
```

```python
plt.figure(figsize=(25,10))
plt.title('Customer Dendogram')
dendrogram(linkage(data.values,method='ward'))
plt.show
```

If we draw a horizontal line that passes through longest distance without a horizontal line, we get 5 clusters.

We create an instance of Agglomerative Clustering using the euclidean distance as the measure of distance between points and ward linkage to calculate the proximity of clusters.

```python
## use now AgglomerativeClustering

from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5, linkage='ward')
cluster.fit_predict(data.values)
```

```python
plt.figure(figsize=(15,8))
plt.title('Digram of Agglomerative clustering')
plt.scatter(data.values[:,0],data.values[:,1],c= cluster.labels_, cmap='rainbow')
```
