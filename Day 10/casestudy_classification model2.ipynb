## using user data findout  which classification model is best accuracy score


```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
```

```python
list_of_accuracies = []
```

```python
df = pd.read_csv('user_data.csv')
```

```python
df.head()
```

```python
x = df.iloc[:,[1,2,3]].values
y = df.iloc[:,-1].values
```

```python
## conver the Categorical to the numerical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:,0] =le.fit_transform(x[:,0])
x
```

```python
y.shape
```

## Train The Model

```python
## Now import the train test model and train the model

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0,test_size=0.20)
```

## Import Linear Regression and r2 Scroe

```python
# Check with the liner regression as this target is multiple then this is a multi-class classification problem

from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = linear.predict(x_test)  # Predictions
r2 = r2_score(y_test, y_pred)
r2
```

```python
list_of_accuracies.append(r2)
```

## Import Logistic Regression and r2 score

```python
# check with the Logistic Regression
from sklearn.linear_model import LogisticRegression
logistic_regression = LogisticRegression()
logistic_regression.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = logistic_regression.predict(x_test)  # Predictions
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
list_of_accuracies.append(accuracy)
```

## Import KNN classifiers

```python
from sklearn.neighbors import KNeighborsClassifier
KNN_Model = KNeighborsClassifier(n_neighbors=5, metric= 'minkowski', p = 2)
KNN_Model.fit(x_train,y_train)
```

```python
y_pred_knn = KNN_Model.predict(x_test)
r2_knn_score = r2_score(y_test,y_pred_knn)
r2_knn_score
```

```python
list_of_accuracies.append(r2_knn_score)
```

## Import SVC Model

```python
from sklearn.svm import SVC
svc_model = SVC(kernel='linear')
svc_model.fit(x_train,y_train)
```

```python
y_pred_svc= svc_model.predict(x_test)
r2_score_svc = r2_score(y_test,y_pred_svc)
r2_score_svc
```

```python
list_of_accuracies.append(r2_score_svc)
```

## Import decissionTree classification

```python
from sklearn.tree import DecisionTreeClassifier
decission_tree_model = DecisionTreeClassifier(max_depth=2,random_state=0)
decission_tree_model.fit(x_train,y_train)
```

```python
y_decision_tree_pred = decission_tree_model.predict(x_test)
r2_decision_tree_score = r2_score(y_test,y_decision_tree_pred)
r2_decision_tree_score
```

```python
list_of_accuracies.append(r2_decision_tree_score)
```

## Import randomForestClassifier


```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(n_estimators=50, random_state=42)
RFC.fit(x_train,y_train)
```

```python
y_pred_RFC = RFC.predict(x_test)
r2_RFC_score = r2_score(y_test,y_pred_RFC)
r2_RFC_score
```

```python
list_of_accuracies.append(r2_RFC_score)
```

## Import Gussian NB

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
y_pred_gnb = gnb.predict(x_test)
r2_gnb_score = r2_score(y_test, y_pred_gnb)
r2_gnb_score
```

```python
list_of_accuracies.append(r2_gnb_score)
```
