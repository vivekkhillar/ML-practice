```python
#import libraries

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('User_Data.csv')
df.head()
```

```python
# pick the dependent and Independent variables

x=df.iloc[:,[2,3]].values
y=df.iloc[:,4].values
```

```python
x
```

```python
y
```

```python
# Now import the Logistic regression model

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
print (train_test_split(x,y, test_size=0.25, random_state=0))
```

```python
# Now upload the Dependent and Independent variables in train test split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
#create an object of the library to use the same
slr = StandardScaler()

# fit the x_train and x_test 

x_train=slr.fit_transform(x_train) # fit_transform
x_test=slr.transform(x_test) # transform
```

```python
x_train
```

```python
x_test
```

```python
lr_model = LogisticRegression()
lr_model.fit(x_train,y_train)
y_pred = lr_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

#### Using KNN classifier 


```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric= 'minkowski', p = 2) 
# n_neighbors i.e. number of data points
# metric 'minkowski' responsible for measure distance
# Measures distances using Euclidean distance (via Minkowski with p=2),
classifier.fit(x_train,y_train) # Model always contains the x_train and y_train during fit
```

```python
y_pred1 = classifier.predict(x_test)
y_pred1
```

```python
accuracy_score(y_test,y_pred1)*100
```

#### Finding the Error in Kn classier using the matplot

```python
error = []

for i in range (1,40):
    knn = KNeighborsClassifier(n_neighbors=i)  #n_neighbors 
    knn.fit(x_train,y_train)
    pred_i = knn.predict (x_test)
    error.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize = (18,6))
plt.plot(range(1,40),error,color = 'red',linestyle='dashed',marker='o',markerfacecolor='blue',markersize=10)

plt.title('Error rate k value')
plt.xlabel('k value')
plt.ylabel('Mean Error')
```
