```python
#### SVM ( SUPPORT VECTOR MACHINE ALGO )

## import the libraries

import os
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot
```

```python
## Import data set from the existing library of sklearn

from sklearn import datasets

cancer = datasets.load_breast_cancer()
```

```python
## Print the Features and lables

print(f"Features: {cancer.feature_names}")
print(f"Lables: {cancer.target_names}")
```

```python
# Now train and split the model

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(cancer.data,cancer.target,test_size=0.3, random_state=109)
```

```python
# import SVM model

from sklearn import svm
svm_model = svm.SVC(kernel='linear')
svm_model.fit(x_train,y_train)
y_pred = svm_model.predict(x_test)
```

```python
# now check the accuracy

from sklearn.metrics import accuracy_score
print (f"Accuracy Score is : {accuracy_score(y_test,y_pred)}")
```

```python
# now print the precission and recall

from sklearn import metrics

print ("Precession:", metrics.precision_score(y_test,y_pred))
print ("Recall Score", metrics.recall_score(y_test,y_pred))
```

```python
# check the confusion metrics

from sklearn.metrics import confusion_matrix
Confusion_Matrix = confusion_matrix(y_test,y_pred)
Confusion_Matrix
```

```python
# Now display the classification report 

from sklearn.metrics import classification_report
print (classification_report(y_test,y_pred))
```

```python
# Now check for IRIS DATA 

iris_data = datasets.load_iris
url="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

```

```python
# specify the col names

col_names = ['sepal-length','sepal-width','petal-length','petal-width','class']
irisdata=pd.read_csv(url,names=col_names)
```

```python
irisdata.head()
```

```python
# check the unique class
irisdata['class'].unique()
```

```python
# Create Lable and Features

x = irisdata.drop('class',axis=1)
y= irisdata['class']
```

```python
# Now again train the model with above feature and lables

X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
# now load the model to the train and testing data

from sklearn.svm import SVC
svc_model = SVC(kernel='poly',degree=1)
svc_model.fit(X_train,Y_train)
Y_pred = svc_model.predict(X_test)
```

```python
print(confusion_matrix(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
```

```python
# Support vector classifier

svclassifier = SVC(kernel='rbf',gamma=0.1)
svclassifier.fit(X_train,Y_train)
```

```python
Y_pred = svclassifier.predict(X_test)
```

```python
print(confusion_matrix(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
```
