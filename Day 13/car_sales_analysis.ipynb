#### CAR sales analysis

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df = pd.read_csv('Car_sales.csv')
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Manufacturer'] = le.fit_transform(df['Manufacturer'])
df['Model'] = le.fit_transform(df['Model'])
df['Vehicle_type'] = le.fit_transform(df['Vehicle_type'])
```

```python
df['Latest_Launch'] = pd.to_datetime(df['Latest_Launch'])

# extract year, month, day from Latest_Launch
df['Launch_year'] = df['Latest_Launch'].dt.year
df['Launch_month'] = df['Latest_Launch'].dt.month
df['Launch_day'] = df['Latest_Launch'].dt.day
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True) 
```

```python
df.info()
```

```python
x = df.drop(columns=['Price_in_thousands','Latest_Launch'])
y = df['Price_in_thousands']
```

```python
x.shape, y.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
# print y_pred and y_test
print("Predicted Prices:", y_pred)
print("Actual Prices:", y_test.values)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
